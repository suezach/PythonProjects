Import the Libraries


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Import the COVID19 Dataset

```python
corona_dataset_csv = pd.read_csv("covid19_Confirmed_dataset.csv")
```

```python
corona_dataset_csv.head()
```

```python
corona_dataset_csv.shape
```

Delete useless columns

```python
corona_dataset_csv.drop(["Lat","Long"], axis = 1, inplace = True)
```

```python
corona_dataset_csv.head(15)
```

Aggregrate the rows by country

```python
corona_dataset_aggregated = corona_dataset_csv.groupby("Country/Region").sum()
corona_dataset_aggregated.head(10)
```

Visualize data related to a country "China"

```python
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["Italy"].plot()
plt.legend()
```

Calculating a good measure

```python
corona_dataset_aggregated.loc["China"][:3].plot()
```

Calculating and plotting the first derivative of the curve

```python
corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
corona_dataset_aggregated.loc["Italy"].diff().plot()
```

```python
corona_dataset_aggregated.loc["Spain"].diff().plot()
```

Calculate the maximum number to find measure that describes the spread of the virus in this data range.
The first derivative ( diff function) is used to find the rate of the curve.

```python
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
corona_dataset_aggregated.loc["Italy"].diff().max()
```

```python
corona_dataset_aggregated.loc["Spain"].diff().max()
```

ANALYSIS: Above we get the maximum number of cases that are seen by China, Italy and Spain in the time period. It is deduced that China has seen the maximum number of cases during the time period studied.

To find the max_infection_rate for all countries


```python
countries = list(corona_dataset_aggregated.index)
max_infection_rate = []
for c in countries:
    max_infection_rate.append(corona_dataset_aggregated.loc[c].diff().max())
max_infection_rate
```

put the above result in a table by creating a new column

```python
corona_dataset_aggregated["max_infection_rate"] = max_infection_rate
```

```python
corona_dataset_aggregated.head(10)
```

Create a new dataframe with only the needed columns of country/region and max_infection_rate

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rate"])
```

```python
corona_data.head()
```

Import new dataset covid_deaths

```python
covid_deaths = pd.read_csv("covid19_deaths_dataset.csv")
```

```python
covid_deaths.head()
```

```python
covid_deaths.drop(["Lat","Long"], axis=1, inplace = True)
```

```python
covid_deaths.head()
```

```python
covid_deaths_aggregated = covid_deaths.groupby("Country/Region").sum()
```

```python
covid_deaths_aggregated.head(15)
```

```python
covid_deaths_aggregated.loc["China"].plot()
```

```python
covid_deaths_aggregated.loc["China"][:3].plot()
```

```python
covid_deaths_aggregated.loc["China"].diff().plot()
```

```python
covid_deaths_aggregated.loc["China"].diff().max()
```

```python
countries = list(covid_deaths_aggregated.index)
cummulative_deaths = []
for c in countries:
    cummulative_deaths.append(covid_deaths_aggregated.loc[c].diff().sum())
cummulative_deaths
```

```python
covid_deaths_aggregated["covid_deaths"] = cummulative_deaths
```

```python
covid_deaths_aggregated.head(40)
```

```python
covid_data_death = pd.DataFrame(covid_deaths_aggregated["covid_deaths"])
```

```python

corona_data.shape
```

```python
corona_data.head()
```

```python
covid_data_death.shape
```

```python
data = corona_data.join(covid_data_death)
```

```python
data.head()
```

```python
data.corr()
```

```python
x = data["covid_deaths"]
y = data["max_infection_rate"]
```

```python
sns.scatterplot(x,y)
```

```python
sns.regplot(x,y)
```

ANALYSIS: There is a positive correlation between the maximum infection rate and covid deaths. As the rate of infection increases, the death rate also increases. So it is not depended on whether the country is developed or not, if enough people are getting tested.
